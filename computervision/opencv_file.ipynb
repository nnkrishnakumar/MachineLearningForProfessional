```python
img=cv2.imread(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\computervision\kate.jpg")  
img2=cv2.resize(img,(160,206))   # kate original image size =1000*1535 px  convert it to 160*206 px  and stored it into a variable called "img2"
cv2.imwrite(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\computervision\resize_kate_image.jpg",img2)
cv2.imshow("win",img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# merge kate image and sachin image together

sachine=cv2.imread(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\computervision\sachine.webp")
kate=cv2.imread(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\computervision\resize_kate_image.jpg")

sachine_kate=cv2.addWeighted(sachine,0.9,kate,0.3,0)
cv2.imshow("window",sachine_kate)
cv2.waitKey()
cv2.destroyAllWindows()
```
