```python
# slicing df1's years of experience into 2D array
X=df1.iloc[:,0:1].values     # this our feature
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
Y_pred_salary=model.predict(X)
```

```python
Y_pred_salary
```

```python
# MSE: Mean Square error 
```

```python
# calculate the error/Loss
Loss=y-Y_pred_salary
```

```python
# Square of error
Loss*Loss
```

```python
Square_loss=np.square(Loss)
Square_loss
```
