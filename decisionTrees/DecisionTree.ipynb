```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\classification_model\IRIS.csv")
```

```python
df
```

```python
X=df.iloc[:,:-1].values
X
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

```python
# threshold values
```

```python
model=DecisionTreeClassifier()
model.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))
plot_tree(model)
plt.show()
```

```python
df.species.value_counts()
```

```python
sample=[[2.5,6.5,4.5,3.2]]
model.predict(sample)
```

```python
model.predict_proba(sample)
```

```python
# notes:

# > decisiontree is overfit most of the time 
# > decisiontree is good for classification not for regression

# How to protect decision tree model to get overfit
# by training not upto end leaf
```

```python
model2=DecisionTreeClassifier(max_depth=3)
model2.fit(X,y)
```

```python
plot_tree(model2)
plt.show()
```
