```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\classification_model\IRIS.csv")
```

```python
df
```

```python
X=df.iloc[:,:-1].values
X
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

```python
# threshold values
```

```python
model=DecisionTreeClassifier(min_samples_leaf=0.4)
model.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```
