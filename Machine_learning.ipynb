```python
fruit
```

```python
fruit.iloc[0:40:1,0:2:1]
```

```python
X=fruit.iloc[:,0:2].values
```

```python
import numpy as np
# Target must be in 1 D array

Y=np.array(fruit.Class)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=11)
```

```python
model.fit(X,Y)
```

```python
model.predict([[69,4]])
```
