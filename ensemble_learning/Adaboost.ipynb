```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.ensemble import AdaBoostClassifier
```

```python
# loading our iris data 
df=pd.read_csv(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\classification_model\IRIS.csv")
df
```

```python
# split data into feature and target

X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#scale your data

from sklearn.preprocessing import MinMaxScaler 
```

```python
sc=MinMaxScaler()
X_sc=sc.fit_transform(X)
X_sc
```

```python
# split data into training and testing part 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,train_size=0.7,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
base_model=LogisticRegression(max_iter=2000,multi_class='multinomial')
model=AdaBoostClassifier(estimator=base_model,n_estimators=100)
model.fit(X_train,y_train)
```

```python
sample=[[5.1,3.5,1.4,0.2]]
sample_sc=sc.transform(sample)
sample_sc
```

```python
model.predict(sample_sc)
```

```python
model.predict_proba(sample_sc)
```
