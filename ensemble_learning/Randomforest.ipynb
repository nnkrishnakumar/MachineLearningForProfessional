```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.ensemble import RandomForestClassifier
df=pd.read_csv(r"C:\Users\aimar\OneDrive\Desktop\Youtube\MachineLearningForProfessional\classification_model\IRIS.csv")
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
#scaling my data using minmaxscaler
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_sc=sc.fit_transform(X)   # fit_transform scaling the data as well as finding min and max values of each column
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,train_size=0.70,random_state=42)
model=RandomForestClassifier(n_estimators=100)
model.fit(X_train,y_train)
sample=[[5.1, 3.5, 1.4, 0.2]]
sample_sc=sc.transform(sample)     # tranform scale the data without calculating min and max values of each column. it is using fit_tranform scale to scale the data sample
print(sample_sc)
print("model prediction for a sample [[5.1, 3.5, 1.4, 0.2]]:",model.predict(sample_sc))
print("probability score:",model.predict_proba(sample_sc))
df.species.value_counts()
df.species.value_counts().index
X_train_pred=model.predict(X_train)
X_test_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
print("accruracy score for training sample:",accuracy_score(X_train_pred,y_train))
print("accuracy score of testing sample:",accuracy_score(X_test_pred,y_test))
```
